```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(rc={'figure.figsize':(16,9)}, palette='husl')
```

```python
learning_dataset = pd.read_csv('target_dataset2.csv.gz')
learning_dataset
```

```python
learning_dataset.info()
```

```python
from sklearn.model_selection import train_test_split

X = learning_dataset.drop(columns=['vmaf_neg_mean'])
y = learning_dataset['vmaf_neg_mean']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.75)
```

```python
X_train
```

```python
X_train.info()
```

```python
y_train
```

```python
from catboost import CatBoostRegressor, Pool, metrics, cv
```

```python
model = CatBoostRegressor(
    custom_metric=[metrics.MAPE(), metrics.SMAPE(), metrics.MAE()],
    logging_level='Silent',
    loss_function='Huber:delta=0.5',
    depth=7,
    iterations=1e5,
    learning_rate=1e-2,
)
model.fit(
    X_train, y_train,
    cat_features=['format_first', 'parameter'],
    eval_set=(X_test, y_test),
#     logging_level='Verbose',  # you can uncomment this for text output
    plot=True
);
```

```python
cv_data = cv(
    Pool(X, y, cat_features=['format_first', 'parameter']),
    dict(
        custom_metric=[metrics.MAPE(), metrics.SMAPE()],
        logging_level='Silent',
        loss_function='RMSE',
        depth=6,
    ),
    plot=True
)
```

```python
y_train.value_counts()
```

```python
model.feature_importances_
```

```python

sns.barplot(
    pd.DataFrame({
        'feature_importance': model.feature_importances_,
        'feature_names': X.columns}).sort_values(
            by=['feature_importance'],
            ascending=False,
    ),
    y='feature_names',
    x='feature_importance'
)
```

```python
learning_dataset_red = pd.read_csv('target_dataset3.csv.gz')
learning_dataset_red
```

```python
from sklearn.model_selection import train_test_split

X_red = learning_dataset_red.drop(columns=['vmaf_neg_mean'])
y_red = learning_dataset_red['vmaf_neg_mean']
X_train, X_test, y_train, y_test = train_test_split(X_red, y_red, train_size=0.75)
```

```python
X_red.sample(10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_red, y_red, train_size=0.75)
model = CatBoostRegressor(
    custom_metric=[metrics.MAPE(), metrics.SMAPE(), metrics.MAE()],
    logging_level='Silent',
    loss_function='Huber:delta=1',
    depth=8,
    iterations=50000,
    learning_rate=1e-2,
)
model.fit(
    X_train, y_train,
    cat_features=['parameter'],
    eval_set=(X_test, y_test),
#     logging_level='Verbose',  # you can uncomment this for text output
    plot=True,
);
```

```python
plt.figure(figsize=(20, 16))
sns.barplot(
    pd.DataFrame({
        'feature_importance': model.feature_importances_,
        'feature_names': X_red.columns}).sort_values(
            by=['feature_importance'],
            ascending=False,
    ),
    y='feature_names',
    x='feature_importance'
)
```

```python
model.save_model('reduced_models.cbm', format='cbm')
```
